```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to pandas dataframe
cc_data = pd.read_csv('/content/creditcard.csv')
```

```python
# getting first 5 rows
cc_data.head()
```

```python
# getting last 5 rows
cc_data.tail()
```

```python
# dataset information
cc_data.info()
```

```python
# checking if any field contains null value
cc_data.isnull().sum()
# data is clean as no null value exists
```

```python
# distribution of legit and fraudulent transactions
#  0 - legit
#  1 - fraud
cc_data['Class'].value_counts()
# distribution is highly unbalanced
```

```python
# separating data for data analysis
legit = cc_data[cc_data.Class==0]
fraud = cc_data[cc_data.Class==1]
```

```python
print('legit data distribution shape : ',legit.shape)
print('fraud data distribution shape : ',fraud.shape)
```

```python
# statistical measures of legit transaction dataset
legit.Amount.describe()
```

```python
# statistical measures of fraud transaction dataset
fraud.Amount.describe()
```

```python
# comparing values for both legit and fraud data
cc_data.groupby('Class').mean()
```

```python
# UNDER SAMPLING
# BUILD A SAMPLE DATASET CONTAINING SIMILAR DISTRIBUTION OF LEGIT AND FRAUD TRANSACTION
```

```python
# NO. OF FRAUD TRANSACTIONS : 492
legit_sample = legit.sample(n=492)
```

```python
# concatenating both the data frames now legit_sample and fraud data
# both of similar distribution now
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
# getting first 5 rows of new dataset
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# splitting dataset into features and target
x = new_data.drop(columns='Class',axis=1)
y = new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , stratify = y , random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
# Model Training
# Logistic Regression
```

```python
regressor = LogisticRegression()
```

```python
# training our model on training dataset
```

```python
regressor.fit(x_train,y_train)
```

```python
# Model Evaluation by Accuracy Score
x_train_predict = regressor.predict(x_train)
train_data_accuracy = accuracy_score(x_train_predict,y_train)
```

```python
print('training data accuracy score : ',train_data_accuracy)
```

```python
# testing our model on test data now
x_test_predict = regressor.predict(x_test)
test_data_accuracy = accuracy_score(x_test_predict,y_test)
```

```python
print('testing data accuracy score : ',test_data_accuracy)
```
